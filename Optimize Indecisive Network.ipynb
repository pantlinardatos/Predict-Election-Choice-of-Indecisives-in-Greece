```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

#From Keras
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils

#From Sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import ExtraTreesClassifier

from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler

from hyperas import optim
from hyperas.distributions import choice, uniform, conditional
from hyperas.distributions import uniform

from hyperopt import Trials, STATUS_OK, tpe
```

```python
def data():
    
    #Load the Data
    x_train = pd.read_csv('x_train.csv', encoding='utf-8').drop('Unnamed: 0',axis=1)
    y_train = pd.read_csv('y_train.csv', encoding='utf-8')['election']
    
    #Make them catecogrical - one hot. That means that if we have 10 classes it will be an array with an 1 at the corresponding pos.
    y_train = np_utils.to_categorical(y_train)
    
    #Normalize 
    scaler = MinMaxScaler()
    fitted_scaler = scaler.fit(x_train)

    x_train = scaler.transform(x_train)
    
    return x_train,y_train
```

```python
def create_model(x_train,y_train):    
    callbacks = [

            keras.callbacks.EarlyStopping(monitor='loss',patience=3),
            keras.callbacks.ModelCheckpoint(filepath='indecisives_model.h5', monitor='val_loss',save_best_only=True),
            keras.callbacks.ReduceLROnPlateau(monitor='val_loss',factor=0.1,patience=1),
            keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=3, verbose=1, mode='auto')
        ]


    #define main model
    model = Sequential()

    model.add(Dense( {{choice([4, 16, 32, 64, 128, 256, 512, 1024])}} , activation='relu', input_shape=(x_train.shape[1],)))
    model.add(Dropout({{uniform(0, 1)}}))

    model.add(Dense( {{choice([4, 16, 32, 64, 128, 256, 512, 1024])}}, activation='relu'))
    model.add(Dropout({{uniform(0, 1)}}))

    model.add(Dense( {{choice([4, 16, 32, 64, 128, 256, 512, 1024])}}, activation='relu'))
    model.add(Dropout({{uniform(0, 1)}}))

    model.add(Dense( {{choice([4, 16, 32, 64, 128, 256, 512, 1024])}}, activation='relu'))
    model.add(Dropout({{uniform(0, 1)}}))

    model.add(Dense( {{choice([4, 16, 32, 64, 128, 256, 512, 1024])}}, activation='relu'))
    model.add(Dropout({{uniform(0, 1)}}))

    model.add(Dense( {{choice([4, 16, 32, 64, 128, 256, 512, 1024])}}, activation='relu'))
    model.add(Dropout({{uniform(0, 1)}}))

    model.add(Dense(y_train.shape[1], activation='softmax'))

    # Compile model
    model.compile(optimizer={{choice(['rmsprop', 'adam', 'sgd'])}} ,loss='categorical_crossentropy',metrics=['accuracy'])

    model.fit(x_train, y_train, epochs=1000, batch_size={{choice([8, 16, 32, 64, 128])}}, 
              validation_split = 0.25,callbacks=callbacks)

    model.load_weights("indecisives_model.h5")
    
    score, acc = model.evaluate(x_train, y_train, verbose=0)
    
    print (acc)
    
    return {'loss': -acc, 'status': STATUS_OK, 'model': model}
```

```python
best_run, best_model = optim.minimize(model=create_model,
                                              data=data,
                                              algo=tpe.suggest,
                                              max_evals=250,
                                              trials=Trials(),notebook_name='Optimize Indecisive Network')

X_train, Y_train = data()
# print("Evalutation of best performing model:")
# print(best_model.evaluate(X_test, Y_test))
print("Best performing model chosen hyper-parameters:")
print(best_run)
print(best_model.evaluate(X_train, Y_train))
```
